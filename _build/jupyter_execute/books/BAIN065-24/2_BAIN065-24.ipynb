# Ecuaciones Algebraicas

Una ecuación es una igualdad entre expresiones algebraicas que contiene una o más cantidades desconocidas denominadas "incógnitas". Se clasifican según el número de incógnitas y el grado de la ecuación (es el mayor exponente que contiene la incógnita).

## Ecuación de Primer Grado

Es una  ecuación de la forma:
$\begin{equation}Ax + B = 0\end{equation}\quad$
(con $A$ y $B$ Números Reales, y $x$ la incógnita).

* Conjunto Solución: Son todos los Números Reales que satisfacen la ecuación.

> $\quad\quad E(x) = 0 \quad;\quad S = \{c\in\mathbb{R} \quad : \quad E(c) = 0\}$

_Obs.:<span style="background-color: yellow"></span>_
En una ecuación podemos realizar las siguientes operaciones sin cambiar el Conjunto Solución.

- Sumar un mismo número a ambos lados de la igualdad.<span style="background-color: rgb(230, 230, 230)"></span>

- Multiplicar un número distinto de cero a ambos lados de la igualdad.<span style="background-color: rgb(230, 230, 230)"></span>

<br></br>
**Ejemplos:** Encuentre el Conjunto Solución.

<span style="background-color: rgb(150, 236, 255)"></span>
> a) $\quad 3x - 1 = 4$
>    
> $\quad\Leftrightarrow 3x - 1 +(1) = 4 + (1)$
>    
> $\quad\Leftrightarrow 3x = 5$
> 
> $\quad\Leftrightarrow 3x \cdot(\frac{1}{3}) = 5 \cdot(\frac{1}{3})$
> 
> $\quad\Leftrightarrow \frac{3x}{3} = \frac{5}{3}$
>
> $\quad\Leftrightarrow x = \frac{5}{3}$
>
> El Conjunto Solución es: <span style="font-size: 12pt">$\quad S = \{\frac{5}{3}\}$</span>

<br></br><span style="background-color: rgb(150, 236, 255)"></span>
> b) $\quad x + 3 = 3\cdot(2 - x) + 4$
>
> $\quad\Leftrightarrow x + 3 = 6 - 3x + 4$
>
> $\quad\Leftrightarrow x + 3 = 6 - 3x + 4$
>
> $\quad\Leftrightarrow x + 3 + 3x = 6 - 3x + 4 + 3x$
>
> $\quad\Leftrightarrow 4x + 3 = 6 + 4$
>
> $\quad\Leftrightarrow 4x + 3 = 10$
>
> $\quad\Leftrightarrow 4x + 3 - 3= 10 - 3$
>
> $\quad\Leftrightarrow 4x = 7$
>
> $\quad\Leftrightarrow 4x \cdot (\frac{1}{4})= 7 \cdot(\frac{1}{4})$
>
> $\quad\Leftrightarrow \frac{4x}{4}= \frac{7}{4}$
>
> $\quad\Leftrightarrow x = \frac{7}{4}$
>
> El Conjunto Solución es: <span style="font-size: 12pt">$\quad S = \{\frac{7}{4}\}$</span>

## Sistemas de Ecuaciones

### Sistemas de 2$\times$2

Un sistema de ecuaciones lineales de 2 incógnitas y 2 ecuaciones es una expresión del tipo:

> $\begin{equation}a_{1} \cdot x + b_{1} \cdot y = c_{1} \quad(1)\nonumber\end{equation}$  
> $\begin{equation}a_{2} \cdot x + b_{2} \cdot y = c_{2} \quad(2)\nonumber\end{equation}$  

> con $\quad a_{1}$, $b_{1}$, $c_{1}$, $a_{2}$, $b_{2}$, $c_{2}\in\mathbb{R}\quad$ y $\quad x$, $y$ las incógnitas.

Se define el par $(m, n)$ como solución del sistema, si este satisface a ambas ecuaciones. 

_Obs.:<span style="background-color: yellow"></span>_
$m$ corresponde al valor de $x$ y $n$ corresponde al valor de $y$.

<br></br>
**Ejemplos:** 

<span style="background-color: rgb(150, 236, 255)"></span>
a) Dado el sistema $\quad\left\{\begin{array}{ll} 3x - 2y = 4 \\ 4x - 3y = 5 \end{array}\right.\quad$ verifque si $(2, 1)$ es solución.

Reemplazando el par $(2, 1)$ en ambas ecuaciones tenemos:

> 1. $\quad\left\{\begin{array}{ll} 3x - 2y = 4 \quad\text{(Ecuación 1)} \\ 4x - 3y = 5 \quad\text{(Ecuación 2)}\end{array}\right.$  
>
> 2. $\quad\left\{\begin{array}{ll} 3x - 2y = 4 \\ 4x - 3y = 5 \end{array}\right. 
\quad\Rightarrow\quad 
\begin{matrix} x = 2 \\ y = 1 \end{matrix}
\quad\equiv\quad 
\left\{\begin{array}{ll} 3 \cdot 2 - 2 \cdot 1 = 4 \\ 4 \cdot 2 - 3 \cdot 1 = 5 \end{array}\right.$  
>
> 3. $\quad\left\{\begin{array}{ll} 6 - 2 = 4 \\ 8 - 3 = 5 \end{array}\right.
\quad\Leftrightarrow\quad
\left\{\begin{array}{ll} 4 = 4 \\ 5 = 5 \end{array}\right.$  

En este ejemplo el par satisface ambas ecuaciones, por lo tanto $(2, 1)$ es solución del sistema.

<br></br>
<span style="background-color: rgb(150, 236, 255)"></span>
b) Dado el sistema $\quad\left\{\begin{array}{ll} 3x - 2y = 4 \\ 4x - 3y = 5 \end{array}\right.\quad$ verifque si $(-2, -1)$ es solución.

Reemplazando el par $(-2, -1)$ en ambas ecuaciones tenemos:

> 1. $\quad\left\{\begin{array}{ll} 3x - 2y = 4 \quad\text{(Ecuación 1)} \\ 4x - 3y = 5 \quad\text{(Ecuación 2)}\end{array}\right.$  
>
> 2. $\quad\left\{\begin{array}{ll} 3x - 2y = 4 \\ 4x - 3y = 5 \end{array}\right. 
\quad\Rightarrow\quad 
\begin{matrix} x = -2 \\ y = -1 \end{matrix}
\quad\equiv\quad 
\left\{\begin{array}{ll} 3 \cdot (- 2) - 2 \cdot (- 1) = 4 \\ 4 \cdot (- 2) - 3 \cdot (- 1) = 5 \end{array}\right.$  
>
> 3. $\quad\left\{\begin{array}{ll} - 6 + 2 = 4 \\ - 8 + 3 = 5 \end{array}\right.
\quad\Leftrightarrow\quad
\left\{\begin{array}{ll} - 4 \neq 4 \\ - 5 \neq 5 \end{array}\right.$  

En este ejemplo el par no satisface ninguna de las ecuaciones, por lo tanto $(-2, -1)$ no es una solución del sistema.

## Métodos de Resolución de Sistemas

### Método de Reducción

Consiste en multiplicar una o ambas ecuaciones por un número tal que una de las incógnitas quede con el mismo coeficiente pero de distinto signo. Luego se suman ambas ecuaciones respetando la igualdad y se resuelve la ecuación resultante.

**Ejemplo:**
> 1. $\quad\left\{\begin{array}{ll} 3x - 2y = 4 \\ 4x - 3y = 5 \end{array}\right.
\quad\Leftrightarrow\quad
\left\{\begin{array}{ll} 3x - 2y = 4 \quad\cdot(3)\\ 4x - 3y = 5 \quad\cdot(- 2)\end{array}\right.$  
>
> 2. $\quad\left\{\begin{array}{ll} (3)\cdot 3x - (3)\cdot 2y = (3)\cdot 4 \\ (- 2)\cdot 4x - (- 2)\cdot 3y = (- 2)\cdot 5 \end{array}\right.$  
>
> 3. $\quad\left\{\begin{array}{ll} + 9x - 6y = 12 \\ - 8x + 6y = - 10 \end{array}\right.\quad
\quad\Leftrightarrow\quad
\quad+ 9x - 8x - 6y + 6y = 12 - 10\quad$  
>
> 4. $\quad x + 0 = 2 \quad\Leftrightarrow\quad x = 2$  

Para hallar $y$, reemplazamos $x = 2$ en cualquiera de las ecuaciones, en este caso en la primera ecuación.

> 5. $\quad 3x - 2y = 4 \quad\Leftrightarrow\quad 3\cdot(2) - 2y = 4 \quad\Leftrightarrow\quad 6 - 2y = 4$  
>
> 6. $\quad - (6) + 6 - 2y = - (6) + 4  \quad\Leftrightarrow\quad - 2y = - 2 \quad\Leftrightarrow\quad \cdot(\frac{1}{2}) - 2y = \cdot(\frac{1}{2}) - 2  \quad\Leftrightarrow\quad - y = - 1 \quad\Leftrightarrow\quad \cdot(- 1) - y = \cdot(- 1) - 1$  
>
> 7. $\quad\Leftrightarrow\quad y = 1$  

> Conjunto Solución: <span style="font-size: 12pt">$\quad S = \{(2, 1)\}$</span>
