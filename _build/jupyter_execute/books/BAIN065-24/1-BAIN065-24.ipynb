# Números Reales y Lenguaje Algebraico

## Axiomas de Cuerpo

Sean $a$ y $b$ Números Reales ($a, b\in\mathbb{R}$), se cumplen las siguientes propiedades:

* **Conmutatividad**
> $a + b = b + a$
> 
> $a \cdot b = b \cdot a$
* **Asociatividad**
> $a + (b + c) = (a + b) + c$
> 
> $a \cdot (b \cdot c) = (a \cdot b) \cdot c$
* **Distributividad**
> $a \cdot (b + c) = a\cdot b + a\cdot c$
* **Neutro**
> $a + 0 = a$
>
> $a \cdot 1 = a$
* **Inverso**
> $a + (-a) = 0$
>
> $a \cdot a^{-1} = 1 \quad;\quad a \neq 0$
* **Clausura**
> $a + b\in\mathbb{R}$
> 
> $a \cdot b\in\mathbb{R}$

### División de Números Enteros

Sean $a$, $b$ Números Enteros, entonces existen $c$ y $r$ Números Enteros tales que:

> $a = b \cdot c + r$

Donde $\quad r = 0\quad$ $\vee\quad r < b$

> $a$: $\quad$Dividendo  
> $b$: $\quad$Divisor  
> $c$: $\quad$Cociente  
> $r$: $\quad$Resto/Residuo

## Números Racionales

El Conjunto de los Números Racionales ($\mathbb{Q}$) está definida de la siguiente manera:

> <span style="font-size: 12pt">$\mathbb{Q} = \{\frac{n}{m} : n, m \in\mathbb{Z}\quad ; \quad m \neq 0 \}$</span>
>
> $n$: $\quad$ Numerador  
> $m$: $\quad$Denominador

### Propiedades

* <span style="font-size: 13pt">$\quad\frac{a}{b}\cdot\frac{c}{d} = \frac{a \cdot c}{b \cdot d}$</span>

* <span style="font-size: 13pt">$\quad\frac{a}{b}\div\frac{c}{d} = \frac{a}{b}\cdot\frac{d}{c} = \frac{a \cdot d}{b \cdot c}$</span>

## Notación Científica

Un Número Real está escrito en Notación Científica si tiene la siguiente forma:

> $a \times 10^{n} \quad,\quad 1 \leq a < 10 \quad\wedge\quad n \in\mathbb{Z}$

**Ejemplos:**

> * $\quad 435792 = 4.35792 \times 10^{5}$  
>
> * $\quad 0.000000322 = 3.22 \times 10{-7}$  
>
> * $\quad 53.7392 = 5.37392 \times 10^{1}$  
>
> * $\quad 0.000000000012 = 1.2 \times 10^{-11}$

## Raiz Enésima

La Raiz Enésima de un Número Real se define como:

> $\sqrt[n]{a}$
>
> $n$ es un Número Natural mayor que 1  
> $a$ es un Número Real

_Obs.:<span style="background-color: yellow"></span>_
* Si $n$ es par, entonces $a \geq 0$
* Si $n$ es impar, entonces $a \in\mathbb{R}$

### Propiedades

> * $\quad\sqrt[n]{a^m} = a^{\frac{m}{n}}$
>
> * $\quad\sqrt[n]{a^n} = a^{\frac{n}{n}} = a \quad;\quad a > 0$
>
> * $\quad\sqrt[n]{a}\cdot\sqrt[n]{b} = \sqrt[n]{a \cdot b} \quad;\quad a, b > 0$
>
> * $\quad\frac{\sqrt[n]{a}}{\sqrt[n]{b}} = \sqrt[n]{\frac{a}{b}} \quad;\quad a, b > 0$
>
> * $\quad(\sqrt[n]{a})^{m} = \sqrt[n]{a^m}$

### Productos Notables

#### Suma por Diferencia y Diferencia de Cuadrados
> $\quad\begin{equation}(a + b)\cdot(a - b) 
 \quad=\quad a^{2} - a\cdot b + b\cdot a - b^{2} 
 \quad=\quad a^{2} - b^{2}\nonumber\end{equation}$

> $\quad\begin{equation}a^{2} - b^{2} 
 \quad=\quad (a + b)\cdot(a - b)\nonumber\end{equation}$

#### Cuadrado de Binomio
> $\quad\begin{equation}(a + b)^{2} 
 \quad=\quad (a + b)\cdot(a + b) 
 \quad\Leftrightarrow\quad a^{2} + 2 a b + b^{2}\nonumber\end{equation}$

> $\quad\begin{equation}(a - b)^{2} 
 \quad=\quad (a - b)\cdot(a - b)
 \quad\Leftrightarrow\quad a^{2} - 2 a b + b^{2}\nonumber\end{equation}$

#### Producto de Binomios con Término Común
> $\quad\begin{equation}(a + b) \cdot (a + c) 
 \quad=\quad a^{2} + (b + c)\cdot a + b \cdot c\nonumber\end{equation}$

#### Cubo de Binomio
> $\quad\begin{equation}(a + b)^{3} \quad=\quad (a + b)\cdot(a + b)\cdot(a + b)\nonumber\end{equation}$
>
> $\quad\begin{equation}\Leftrightarrow\quad (a + b)\cdot(a^{2} + 2ab + b^{2})\quad\Leftrightarrow\quad a^{3} + 3 a^{2} b + 3 a b^{2} + b^{3}\nonumber\end{equation}$

> $\quad\begin{equation}(a - b)^3 \quad=\quad (a - b)\cdot(a - b)\cdot(a - b)\nonumber\end{equation}$
> 
> $\quad\begin{equation}\Leftrightarrow\quad (a - b)\cdot(a^{2} - 2ab + b^{2})\quad\Leftrightarrow\quad a^{3} - 3a^{2} b + 3ab^{2} - b^{3}\nonumber\end{equation}$
