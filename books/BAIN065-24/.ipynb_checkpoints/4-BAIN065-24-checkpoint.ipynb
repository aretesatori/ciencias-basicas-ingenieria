# Lógica y Conjuntos

## Proposición

Definición: Una proposición es cualquier frase o afirmación que es siempre verdadera o siempre es falsa.

Ejemplo:

a) $2 + 2 = 4\quad$  

b) $2 + 2 = 5\quad$  

c) $\sqrt{16} = \pm 4$  

d) $\sqrt{25 + 36} = \sqrt{25} + \sqrt{36}$

## Tablas de Verdad

### Conjunción
* $\land$ : Es verdadero cuando todas las proposiciones que la componen son verdaderas.

| $p$ | $q$ | $p \land q$ |
| :-: | :-: | :---------: |
| $V$ | $V$ |     $V$     |
| $V$ | $F$ |     $F$     |
| $F$ | $V$ |     $F$     |
| $F$ | $F$ |     $F$     |

### Disyunción
* $\lor$ : Es verdadero si al menos uno de los valores de verdad es verdadero.

| $p$ | $q$ | $p \lor q$ |
| :-: | :-: | :--------: |
| $V$ | $V$ |    $V$     |
| $V$ | $F$ |    $V$     |
| $F$ | $V$ |    $V$     |
| $F$ | $F$ |    $F$     |

### Implicancia
* $\Rightarrow$ : Es falso cuando el antecedente es verdadero y el consecuente es falso. En todos los otros casos, es verdadero.

| $p$ | $q$ | $p \Rightarrow q$ |
| :-: | :-: | :---------------: |
| $V$ | $V$ |        $V$        |
| $V$ | $F$ |        $F$        |
| $F$ | $V$ |        $V$        |
| $F$ | $F$ |        $V$        |

### Equivalencia
* $\Leftrightarrow$ : Es verdadero si al menos uno de los valores de verdad es verdadero.

| $p$ | $q$ | $p \Leftrightarrow q$ |
| :-: | :-: | :-------------------: |
| $V$ | $V$ |          $V$          |
| $V$ | $F$ |          $F$          |
| $F$ | $V$ |          $F$          |
| $F$ | $F$ |          $V$          |
