## Producto Punto

Sean $\overrightarrow{\rm u}$, $\overrightarrow{\rm v}$ $\in \mathbb{R}^n$, se define el **Producto Punto** entre $\overrightarrow{\rm u}$, $\overrightarrow{\rm v}$ denotado por $\overrightarrow{\rm u} \bullet \overrightarrow{\rm v}$ como:

$$\overrightarrow{\rm u} \bullet \overrightarrow{\rm v} \quad=\quad \sum_{i = 1}^n (\rm u_i \cdot \rm v_i) \quad=\quad ((\rm u_1 \cdot \rm v_1) + (\rm u_2 \cdot \rm v_2) + (\rm u_3 \cdot \rm v_3) + (...) + (\rm u_n \cdot \rm v_n))$$

Ejemplo:

Sean $\overrightarrow{\rm u} = (1, 2, 3)$; $\overrightarrow{\rm v} = (6, 5, 4)$.  

$\overrightarrow{\rm u} \bullet \overrightarrow{\rm v} = (1 \cdot 6 + 2 \cdot 5 + 3 \cdot 4) = (6 + 10 + 12) = 28$

### Propiedades
Sean $\overrightarrow{\rm u}$, $\overrightarrow{\rm v}$, $\overrightarrow{\rm w}$ $\in \mathbb{R}^n$, $k \in \mathbb{R}$, se cumplen las siguientes propiedades.

* $\overrightarrow{\rm u} \bullet \overrightarrow{\rm v} = \overrightarrow{\rm v} \bullet \overrightarrow{\rm u}$

* $\overrightarrow{\rm u} \bullet (\overrightarrow{\rm v} + \overrightarrow{\rm w}) = \overrightarrow{\rm u} \bullet \overrightarrow{\rm v} + \overrightarrow{\rm u} \bullet \overrightarrow{\rm w}$


* $\overrightarrow{\rm 0} \bullet \overrightarrow{\rm v} = 0$

* $\overrightarrow{\rm u} \bullet \overrightarrow{\rm u} = (\|\overrightarrow{\rm u}\|)^2$

## Producto Cruz

Sean $\overrightarrow{\rm u} = \rm(u_1, u_2, u_3), \overrightarrow{\rm v} = (v_1, v_2, v_3) \in \mathbb{R}^3$.  

$$\boxed{\overrightarrow{\rm u} \times \overrightarrow{\rm v} = \rm(u_2 \cdot v_3 - v_2 \cdot u_3, v_1 \cdot u_2 - u_1 \cdot v_3, u_1 \cdot v_2 - v_1 \cdot u_2)}$$

### Propiedades

Sean $\overrightarrow{\rm u}$, $\overrightarrow{\rm v}$, $\overrightarrow{\rm w}$ $\in \mathbb{R}^3$, $k \in \mathbb{R}$, se cumplen las siguientes propiedades.

* $\overrightarrow{\rm u} \times \overrightarrow{\rm w} = -\overrightarrow{\rm w} \times \overrightarrow{\rm u}$  

* $\overrightarrow{\rm u} \bullet (\overrightarrow{\rm u} \times \overrightarrow{\rm w}) = \overrightarrow{\rm w} \bullet (\overrightarrow{\rm u} \times \overrightarrow{\rm w})$

# Resumen de Geometría Vectorial

## Definiciones Básicas

### Vector

Un **vector** es una cantidad que tiene tanto magnitud como dirección. En el plano o en el espacio, un vector se puede representar como un segmento dirigido.

- **Notación**: Un vector en $\mathbb{R}^n$ se denota típicamente como $\vec{v}$ o $\mathbf{v}$.
- **Componentes**: En $\mathbb{R}^2$, un vector $\vec{v}$ se puede escribir como $\vec{v} = \langle v_1, v_2 \rangle$. En $\mathbb{R}^3$, se escribe como $\vec{v} = \langle v_1, v_2, v_3 \rangle$.

### Magnitud de un Vector

La **magnitud** o **norma** de un vector $\vec{v} = \langle v_1, v_2 \rangle$ en $\mathbb{R}^2$ se define como:

$ |\vec{v}| = \sqrt{v_1^2 + v_2^2} $

Para un vector $\vec{v} = \langle v_1, v_2, v_3 \rangle$ en $\mathbb{R}^3$:

$ |\vec{v}| = \sqrt{v_1^2 + v_2^2 + v_3^2} $

### Suma de Vectores

La **suma** de dos vectores $\vec{u} = \langle u_1, u_2 \rangle$ y $\vec{v} = \langle v_1, v_2 \rangle$ en $\mathbb{R}^2$ se define como:

$ \vec{u} + \vec{v} = \langle u_1 + v_1, u_2 + v_2 \rangle $

### Producto por un Escalar

El **producto** de un vector $\vec{v} = \langle v_1, v_2 \rangle$ por un escalar \(c\) es:

$ c\vec{v} = \langle cv_1, cv_2 \rangle $

## Propiedades de los Vectores

1. **Conmutativa de la Suma**:
   $ \vec{u} + \vec{v} = \vec{v} + \vec{u} $

2. **Asociativa de la Suma**:
   $ (\vec{u} + \vec{v}) + \vec{w} = \vec{u} + (\vec{v} + \vec{w}) $

3. **Elemento Neutro**:
   $ \vec{v} + \vec{0} = \vec{v} $

4. **Elemento Inverso**:
   $ \vec{v} + (-\vec{v}) = \vec{0} $

5. **Distribución del Producto Escalar sobre la Suma de Vectores**:
   $ c(\vec{u} + \vec{v}) = c\vec{u} + c\vec{v} $

6. **Distribución del Producto Escalar sobre la Suma de Escalares**:
   $ (c + d)\vec{v} = c\vec{v} + d\vec{v} $

7. **Asociativa del Producto Escalar**:
   $ c(d\vec{v}) = (cd)\vec{v} $

## Producto Escalar

El **producto escalar** (o producto punto) de dos vectores $\vec{u} = \langle u_1, u_2 \rangle$ y $\vec{v} = \langle v_1, v_2 \rangle$ en $\mathbb{R}^2$ es:

$ \vec{u} \cdot \vec{v} = u_1v_1 + u_2v_2 $

En $\mathbb{R}^3$, para $\vec{u} = \langle u_1, u_2, u_3 \rangle$ y $\vec{v} = \langle v_1, v_2, v_3 \rangle$:

$ \vec{u} \cdot \vec{v} = u_1v_1 + u_2v_2 + u_3v_3 $

El producto escalar tiene las siguientes propiedades:

1. **Conmutativa**:
   $ \vec{u} \cdot \vec{v} = \vec{v} \cdot \vec{u} $

2. **Distribución**:
   $ \vec{u} \cdot (\vec{v} + \vec{w}) = \vec{u} \cdot \vec{v} + \vec{u} \cdot \vec{w} $

3. **Asociativa con el Producto por un Escalar**:
   $ (c\vec{u}) \cdot \vec{v} = c(\vec{u} \cdot \vec{v}) $

4. **Producto Escalar de un Vector Consigo Mismo**:
   $ \vec{v} \cdot \vec{v} = |\vec{v}|^2 $

## Producto Vectorial

El **producto vectorial** de dos vectores $\vec{u} = \langle u_1, u_2, u_3 \rangle$ y $\vec{v} = \langle v_1, v_2, v_3 \rangle$ en $\mathbb{R}^3$ es un vector definido como:

$ \vec{u} \times \vec{v} = \langle (u_2 v_3 - u_3 v_2), (u_3 v_1 - u_1 v_3), (u_1v_2 - u_2v_1) \rangle $

El producto vectorial tiene las siguientes propiedades:

1. **Anticonmutativa**:
   $ \vec{u} \times \vec{v} = -(\vec{v} \times \vec{u}) $

2. **Distribución**:
   $ \vec{u} \times (\vec{v} + \vec{w}) = \vec{u} \times \vec{v} + \vec{u} \times \vec{w} $

3. **Asociativa con el Producto por un Escalar**:
   $ (c\vec{u}) \times \vec{v} = c(\vec{u} \times \vec{v}) $

## Ejemplos

### Ejemplo de Suma de Vectores

Dados los vectores $\vec{u} = \langle 1, 3 \rangle$ y $\vec{v} = \langle 4, -2 \rangle$:

$ \vec{u} + \vec{v} = \langle 1 + 4, 3 - 2 \rangle = \langle 5, 1 \rangle $

### Ejemplo de Producto Escalar

Dados los vectores $\vec{u} = \langle 1, 2, 3 \rangle$ y $\vec{v} = \langle 4, -5, 6 \rangle$:

$ \vec{u} \cdot \vec{v} = (1)(4) + (2)(-5) + (3)(6) = 4 - 10 + 18 = 12 $

### Ejemplo de Producto Vectorial

Dados los vectores \(\vec{u} = \langle 1, 2, 3 \rangle\) y \(\vec{v} = \langle 4, -5, 6 \rangle\):

$ \vec{u} \times \vec{v} = \langle (2)(6) - (3)(-5), (3)(4) - (1)(6), (1)(-5) - (2)(4) \rangle = \langle 12 + 15, 12 - 6, -5 - 8 \rangle = \langle 27, 6, -13 \rangle $
