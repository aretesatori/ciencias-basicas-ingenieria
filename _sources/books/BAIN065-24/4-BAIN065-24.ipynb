# Lógica y Conjuntos

## Proposición

**Definición:** Una proposición es una frase o afirmación que posee un único valor de verdad, es decir, Verdadero ($V$) o Falso ($F$).

**Ejemplos:**  

$a) \quad 2 + 2 = 4\quad \text{(Verdadero)}$  

$b) \quad 2 + 2 = 5\quad \text{(Falso)}$   

$c) \quad\sqrt{16} = \pm 4\quad \text{(Falso)}$   

$d) \quad\sqrt{25 + 36} = \sqrt{25} + \sqrt{36}\quad \text{(Falso)}$

## Tablas de Verdad

### Conjunción
* $\land$ : Es verdadero cuando todas las proposiciones que la componen son verdaderas.

| $p$ | $q$ | $p \land q$ |
| :-: | :-: | :---------: |
| $V$ | $V$ | $V$         |
| $V$ | $F$ | $F$         |
| $F$ | $V$ | $F$         |
| $F$ | $F$ | $F$         |

### Disyunción
* $\lor$ : Es verdadero si al menos uno de los valores de verdad es verdadero.

| $p$ | $q$ | $p \lor q$ |
| :-: | :-: | :--------: |
| $V$ | $V$ | $V$        |
| $V$ | $F$ | $V$        |
| $F$ | $V$ | $V$        |
| $F$ | $F$ | $F$        |

### Implicancia
* $\Rightarrow$ : Es falso cuando el antecedente es verdadero y el consecuente es falso. En todos los otros casos, es verdadero.

| $p$ | $q$ | $p \Rightarrow q$ |
| :-: | :-: | :---------------: |
| $V$ | $V$ | $V$               |
| $V$ | $F$ | $F$               |
| $F$ | $V$ | $V$               |
| $F$ | $F$ | $V$               |

### Equivalencia
* $\Leftrightarrow$ : Es verdadero si al menos uno de los valores de verdad es verdadero.

| $p$ | $q$ | $p \Leftrightarrow q$ |
| :-: | :-: | :-------------------: |
| $V$ | $V$ | $V$                   |
| $V$ | $F$ | $F$                   |
| $F$ | $V$ | $F$                   |
| $F$ | $F$ | $V$                   |

### Leyes de Morgan

* $\overline{p \land q} \quad\equiv\quad \overline{p}\lor\overline{q}$
* $\overline{p \lor q} \quad\equiv\quad \overline{p}\land\overline{q}$

### Equivalencias de la Implicancia

* $(p \Rightarrow q) \Leftrightarrow (\overline{p} \lor q)$
* $(p \Rightarrow q) \Leftrightarrow (\overline{q} \Rightarrow \overline{p})$

### Leyes de Absorción

* $[p \land (p \lor q)] \equiv p$
* $[p \lor (p \land q)] \equiv p$

## Intervalos

### Intervalo Aberto

$$]a, b[ \quad=\quad \{x \in \mathbb{R}: a < x < b\}$$

### Intervalo Cerrado

$$[a, b] \quad=\quad \{x \in \mathbb{R}: a \leq x \leq b\}$$

### Intervalos Semiabiertos

## Operatoria de Conjuntos

### Intersección

$$A \cap B = \{x \in \mathbb{U} : x \in A \land x \in B\}$$

### Unión

$$A \cup B = \{x \in \mathbb{U} : x \in A \lor x \in B\}$$

### Complemento

$$A^C = \{x \in \mathbb{U} : x \notin A\}$$

### Diferencia

$$A - B = \{x \in \mathbb{U} : x \in A \land x \notin B\}$$
